```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
path = '../Imagens/'
plt.rcParams['image.cmap'] = 'gray' # Para mostrar as imagens em escala de cinza
plt.rcParams["figure.dpi"] = 100 
```

```python
print('Using ' + cv2.__name__ + ' version ' + cv2.__version__)
```

```python
def imshow(img):
    cv2.namedWindow('img', cv2.WINDOW_KEEPRATIO)
    cv2.imshow('img', img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

## Carregando imagem em escala de cinza

```python
imfile = 'm3gtr.jpg'
img_gray = cv2.imread(os.path.join(path, imfile), cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray)
plt.show()
#Guarde essa imagem numa varivel
```

# Carregando imagem colorida

```python
imfile = 'm3gtr.jpg'
img_color = cv2.imread(os.path.join(path,imfile))
plt.imshow(cv2.cvtColor(img_color, cv2.COLOR_BGR2RGB))
plt.show()
```

Analisando os tipos de dados e convertendo entre eles (uint8, uint16, double, etc.)

```python
#Analisando os tipos de dados e convertendo entre eles (uint8, uint16, double, etc.)
print('img_gray.dtype = ' + str(img_gray.dtype))
print('img_color.dtype = ' + str(img_color.dtype))
print('img_gray.shape = ' + str(img_gray.shape))
print('img_color.shape = ' + str(img_color.shape))
print('img_gray.min() = ' + str(img_gray.min()))
print('img_gray.max() = ' + str(img_gray.max()))
print('img_color.min() = ' + str(img_color.min()))
print('img_color.max() = ' + str(img_color.max()))
print('img_gray.mean() = ' + str(img_gray.mean()))
print('img_color.mean() = ' + str(img_color.mean()))

#Convertendo para uint8
img_gray_uint8 = np.uint8(img_gray)
img_color_uint8 = np.uint8(img_color)

#Convertendo para uint16
img_gray_uint16 = np.uint16(img_gray)
img_color_uint16 = np.uint16(img_color)

#Convertendo para double
img_gray_double = np.double(img_gray)
img_color_double = np.double(img_color)

#Convertendo para float
img_gray_float = np.float32(img_gray)
img_color_float = np.float32(img_color)

#Convertendo para int
img_gray_int = np.int32(img_gray)
img_color_int = np.int32(img_color)



```

Realizando operaes aritmticas entre duas matrizes de tipos de dados diferentes e avaliando o que acontece.

```python
#Realizando operaes aritmticas entre duas matrizes de tipos de dados diferentes e avaliando o que acontece.
def matrix_operations_gray(img1, img2):
    #tipo das imagens
    print('img1.dtype = ' + str(img1.dtype))
    print('img2.dtype = ' + str(img2.dtype))

    #soma
    print("\nSoma:")
    img_sum = img1 + img2
    print('img_sum.dtype = ' + str(img_sum.dtype))
    print('img_sum.min() = ' + str(img_sum.min()))
    print('img_sum.max() = ' + str(img_sum.max()))
    print('img_sum.mean() = ' + str(img_sum.mean()))
    plt.imshow(img_sum)
    plt.show()

    #subtrao
    print("\nSubtrao:")
    img_sub = img1 - img2
    print('img_sub.dtype = ' + str(img_sub.dtype))
    print('img_sub.min() = ' + str(img_sub.min()))
    print('img_sub.max() = ' + str(img_sub.max()))
    print('img_sub.mean() = ' + str(img_sub.mean()))
    plt.imshow(img_sub)
    plt.show()

    #multiplicao
    print("\nMultiplicao:")
    img_mult = img1 * img2
    print('img_mult.dtype = ' + str(img_mult.dtype))
    print('img_mult.min() = ' + str(img_mult.min()))
    print('img_mult.max() = ' + str(img_mult.max()))
    print('img_mult.mean() = ' + str(img_mult.mean()))
    plt.imshow(img_mult)
    plt.show()

    #diviso
    print("\nDiviso:")
    #verificando se h diviso por zero
    if img2.min() == 0:
        #somar pequeos valores para evitar diviso por zero
        img2  = img2 + 1e-6
    img_div = img1 / img2
    print('img_div.dtype = ' + str(img_div.dtype))
    print('img_div.min() = ' + str(img_div.min()))
    print('img_div.max() = ' + str(img_div.max()))
    print('img_div.mean() = ' + str(img_div.mean()))
    plt.imshow(img_div)
    plt.show()
    
```

```python
def matrix_operations_color(img1, img2):
    #tipo das imagens
    print('img1.dtype = ' + str(img1.dtype))
    print('img2.dtype = ' + str(img2.dtype))

    #soma
    print("\nSoma:")
    img_sum = img1 + img2
    print('img_sum.dtype = ' + str(img_sum.dtype))
    print('img_sum.min() = ' + str(img_sum.min()))
    print('img_sum.max() = ' + str(img_sum.max()))
    print('img_sum.mean() = ' + str(img_sum.mean()))
    plt.imshow(cv2.cvtColor(img_sum, cv2.COLOR_BGR2RGB))
    plt.show()

    #subtrao
    print("\nSubtrao:")
    img_sub = img1 - img2
    print('img_sub.dtype = ' + str(img_sub.dtype))
    print('img_sub.min() = ' + str(img_sub.min()))
    print('img_sub.max() = ' + str(img_sub.max()))
    print('img_sub.mean() = ' + str(img_sub.mean()))
    plt.imshow(cv2.cvtColor(img_sub, cv2.COLOR_BGR2RGB))
    plt.show()

    #multiplicao
    print("\nMultiplicao:")
    img_mult = img1 * img2
    print('img_mult.dtype = ' + str(img_mult.dtype))
    print('img_mult.min() = ' + str(img_mult.min()))
    print('img_mult.max() = ' + str(img_mult.max()))
    print('img_mult.mean() = ' + str(img_mult.mean()))
    plt.imshow(cv2.cvtColor(img_mult, cv2.COLOR_BGR2RGB))
    plt.show()

    #diviso

    print("\nDiviso:")
    #verificando se h diviso por zero
    if img2.min() == 0:
        img2 = img2 + 1e-6
   
    img_div = img1 / img2
    print('img_div.dtype = ' + str(img_div.dtype))
    if img_div.dtype == np.float64:
        print("entrou")
        img_div = np.uint8(img_div)
    print('img_div.dtype = ' + str(img_div.dtype))
    print('img_div.min() = ' + str(img_div.min()))
    print('img_div.max() = ' + str(img_div.max()))
    print('img_div.mean() = ' + str(img_div.mean()))
    plt.imshow(cv2.cvtColor(img_div, cv2.COLOR_BGR2RGB))
    plt.show()
```

```python
print("Analisando operaes aritmticas entre matrizes uint8 e uint16 em escala de cinza")
matrix_operations_gray(img_gray_uint8, img_gray_uint16)

print("\nAnalisando operaes aritmticas entre matrizes uint8 e double em escala de cinza")
matrix_operations_gray(img_gray_uint8, img_gray_double)

print("\nAnalisando operaes aritmticas entre matrizes uint8 e float em escala de cinza")
matrix_operations_gray(img_gray_uint8, img_gray_float)

print("\nAnalisando operaes aritmticas entre matrizes uint8 e int em escala de cinza")
matrix_operations_gray(img_gray_uint8, img_gray_int)

print("\nAnalisando operaes aritmticas entre matrizes uint8 e uint16 coloridas")
matrix_operations_color(img_color_uint8, img_color_uint16)

#print("\nAnalisando operaes aritmticas entre matrizes uint8 e double coloridas")
#matrix_operations_color(img_color_uint8, img_color_double)

print("\nAnalisando operaes aritmticas entre matrizes uint8 e float coloridas")
matrix_operations_color(img_color_uint8, img_color_float)

#print("\nAnalisando operaes aritmticas entre matrizes uint8 e int coloridas")
#matrix_operations_color(img_color_uint8, img_color_int)



```

### Resultados
- Em muitos casos, as operaes entre tipos de dados diferentes podem resultar em matrizes de tipo mais abrangente para acomodar uma faixa maior de valores, como float64.
- Quando a matriz resultante  do tipo uint16, valores abaixo de zero so representados como 0.
- No geral, nas imagens em escala de cinza, operaes de soma e multiplicao resultaram numa maior saturao e intensidade da cor preta, respectivamente.
- Nas imagens coloridas, a soma tambm aumentou a saturao, enquanto que a multiplicao aumentou a intensidade da cor branca.
- Na subtrao, as imagens ficaram totalmente escuras, pois os valores negativos foram representados como 0.
